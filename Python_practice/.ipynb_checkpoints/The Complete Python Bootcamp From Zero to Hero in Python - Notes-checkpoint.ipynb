```python
# Interactions between Python Functions

from random import shuffle

example = [4,6,8,2,3,1,6,7]
shuffle(example)
example

# Game Problem: 
def shuffle_list(mylist):
    shuffle(mylist)
    return mylist

result = shuffle_list(example)
result

def player_guess():
    guess=''
    guess= input()
    
    while guess not in ['0','1','2']:
        guess= input("Pick a number:")
    
    return int(guess)
    
def check_guess(mylist,guess):
    if mylist[guess] == '0':
        print('Correct')
    else:
        print('Wrong guess!')
        print(mylist)
        
# Initial list:
mylist = [' ', 'o', '']

# Shuffle list:
mixedup_list = shuffle_list(mylist)

# User Guess:
guess = player_guess()

# Check Guess:
check_guess(mixedup_list, guess)
```

```python
# *args and **kwargs in Python

# def myfunc(a,b):
#     # Returns 5% of the sum of a and b:
#     return sum((a,b)) * 0.05

# myfunc(10,20)
# myfunc(10,20,30)

# *args:
def myfunc(*args):
    return sum(args) * 0.05

myfunc(10,20,30,40,50,80)

# *kwargs:
def myfunc(**kwargs):
    if 'fruit' in kwargs:
        print('my fruit of choice is {}'.format(kwargs))
    else:
        print("I didn't find any fruit here")

myfunc(fruit='apple', veggie='lettuce')

# def myfunc(*args, **kwargs):
#     print(args)
#     print(kwargs)
#     print("I would like {}.{}".format(args[0], kwargs))

# myfunc(10,20,30,fruit='orange', food='eggs', animal='dog',100)
```

```python
# Lambda Expressions, Map, and Filter Functions

# map function:
def square(num):
    return num**2

mynums = [5,3,4,1,2]
for item in map(square, sorted(mynums)):
    print(item)

print(list(map(square, sorted(mynums))))

def splicer(mystring):
    if len(mystring) % 2 == 0:
        return 'Even'
    else:
        return mystring[0]
    
names = ['Swapnadeep', 'Susmita', 'Ishani']

list(map(splicer, names))

# filter function:
def checkeven(num):
    return num%2 == 0

mynums = [5,3,4,1,2]
list(filter(checkeven,mynums))

for n in filter(checkeven,mynums):
    print(n) 

# square function:
def square(num):
    return num**2

square(3)

# lambda fucntion:
square = lambda num: num**2
square(5)

print(list(map(lambda num: num**2, mynums)))
print(list(filter(lambda num: num%2==0, mynums)))

# 
names
# print(list(map(lambda names: names[0],mynums)))
```

```python
# Nested Statements and Scope:

x = 25

def printer():
    x = 50
    return x

print(x)
print(printer())

# Python always followes LEGB rule:
# Example:

lambda num: num**2

# Global:
name = 'This is a Global String'
def greet():    
    # Enclosing:
    name = 'Sammy'    
    def hello():
        # Local:
        name = "I am in Local testing"
        print("Hello " + name)        
    hello()    
greet()

x  = 50

def func(x):
    global x
    print(f"X is {x}")
    
    # Local re-assignemnt on global variable!
    x = "New Val"
    print(f'Just changed the value locally')
    return x

# print(x)
x = func(x)
x
```
